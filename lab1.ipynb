```python
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate random list of size n
n = 10000
arr = random.sample(range(1, n*10), n)

# Best Case: Element at the start
key_best = arr[0]
start_time = time.time()
linear_search(arr, key_best)
time_best = (time.time() - start_time) * 1000  # milliseconds

# Average Case: Element in the middle
key_avg = arr[n // 2]
start_time = time.time()
linear_search(arr, key_avg)
time_avg = (time.time() - start_time) * 1000  # milliseconds

# Worst Case: Element not present
key_worst = -1  # Guaranteed not in array
start_time = time.time()
linear_search(arr, key_worst)
time_worst = (time.time() - start_time) * 1000  # milliseconds

# Plotting
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [time_best, time_avg, time_worst]

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.title('Linear Search Time Analysis (n = 10000)')
plt.xlabel('Case Type')
plt.ylabel('Time Taken (ms)')
for i, t in enumerate(times):
    plt.text(i, t + 0.1, f'{t:.3f} ms', ha='center')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print times
print(f"Best Case Time: {time_best:.3f} ms")
print(f"Average Case Time: {time_avg:.3f} ms")
print(f"Worst Case Time: {time_worst:.3f} ms")
```

```python
import random
import matplotlib.pyplot as plt

# Linear search that counts comparisons
def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for item in arr:
        comparisons += 1
        if item == key:
            break
    return comparisons

# Input sizes to test
input_sizes = [100, 500, 1000, 5000, 10000]

# Store comparison counts
comparisons_best = []
comparisons_avg = []
comparisons_worst = []

for n in input_sizes:
    arr = random.sample(range(1, n*10), n)  # Unique random integers

    # Best case: key at the beginning
    key_best = arr[0]
    comp_best = linear_search_with_comparisons(arr, key_best)

    # Average case: key in the middle
    key_avg = arr[n // 2]
    comp_avg = linear_search_with_comparisons(arr, key_avg)

    # Worst case: key not present
    key_worst = -1  # Guaranteed to not be in the array
    comp_worst = linear_search_with_comparisons(arr, key_worst)

    comparisons_best.append(comp_best)
    comparisons_avg.append(comp_avg)
    comparisons_worst.append(comp_worst)

# Plotting
plt.plot(input_sizes, comparisons_best, marker='o', label='Best Case (O(1))')
plt.plot(input_sizes, comparisons_avg, marker='o', label='Average Case (O(n/2))')
plt.plot(input_sizes, comparisons_worst, marker='o', label='Worst Case (O(n))')
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.title("Linear Search: Comparisons vs Input Size")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Tabulated Output
print("Input Size | Best Case | Average Case | Worst Case")
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:10} | {comparisons_best[i]:9} | {comparisons_avg[i]:13} | {comparisons_worst[i]:11}")
```
